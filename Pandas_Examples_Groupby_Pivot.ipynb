```python
orders_db = df.copy()

crosstab_pivot=pd.crosstab(orders_db.CustomerID,orders_db.Product,values=orders_db.Quantity,aggfunc='sum',
                          margins=True,margins_name='Total_Qty').fillna(0)
crosstab_pivot
```

```python
crosstab_pivot = pd.crosstab(
    [orders_db.CustomerID,
    orders_db.Date,
    orders_db.Address],
    [orders_db.Product],
    values = orders_db.Quantity,
    aggfunc='sum', margins=True, margins_name='Total Qty').fillna(0)

pd.set_option('max_columns',100)
crosstab_pivot
```

```python
orders_db.groupby(['CustomerID','Date','Address','Product']).agg({'Quantity':'sum'})
```

```python
pd.pivot_table(orders_db, values='Quantity',
              index=['CustomerID', 'Address'],
              columns='Product',
              aggfunc=['sum',len,'mean'],
              margins=True,fill_value=0)
```

```python
orders_db.groupby(['CustomerID','Address','Product']).agg({'Quantity':['sum',len,'mean']})
```

```python
big_orders = orders_db.groupby(['CustomerID','Product']).filter(lambda x: x['Quantity'].sum()>3)
big_orders
```

```python
orders_db.groupby(['Name','Product']).agg({'Quantity':'sum'})
```
